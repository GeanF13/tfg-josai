```python
def normalize_text(text):
    text = text.lower()
    #text = re.sub(r"[^a-záéíóúñü0-9\s.,:;\-\(\)]", "", text)
    return text

pages = [{'page_number':page['page_number'], 'content': normalize_text(page['content'])} for page in pages]
```

```python
def remove_headers(text):
    words_to_eliminate = ["pr", "proceso", "coordinación", "coordinacion", "enseñanzas", "anx", "cl", "guía", "guia", "aprendizaje", "e.t.s", "sistemas"]
    
    lines = text.splitlines()
    max_lines = min(7, len(lines))
    
    for i in range(max_lines):
        if any(word in lines[i] for word in words_to_eliminate):
            lines[i] = ""
    
    text = "\n".join(line for line in lines if line.strip() != "")
    return text

for i in range(1, len(pages)):
    pages[i]['content'] = remove_headers(pages[i]['content'])
#pages = [{'page_number':page['page_number'], 'content': remove_headers(page['content'])} for page in pages]
```

```python
def remove_footers(text):
    
    lines = text.splitlines()
    total_lines = len(lines)
    
    if total_lines < 5:
        return text
    
    idx_first = total_lines - 5
    idx_fourth = total_lines - 4
    idx_third = total_lines - 3
    idx_second = total_lines - 2
    idx_last = total_lines - 1
    
    pattern_line1 = re.compile(r"^ga_61.._\d+$", re.IGNORECASE)
    pattern_line2 = re.compile(r"^\d+[a-z]?_\d{4}-\d{2}$", re.IGNORECASE)
    pattern_line5 = re.compile(r"^p[áa]gina\s+\d+\s+de\s+\d+$", re.IGNORECASE)
    
    removed_line1 = False
    removed_line2 = False
    removed_line5 = False
    
    if pattern_line1.match(lines[idx_first]):
        lines[idx_first] = ""
        removed_line1 = True
        
    if pattern_line2.match(lines[idx_fourth]):
        lines[idx_fourth] = ""
        removed_line2 = True
        
    if pattern_line5.match(lines[idx_last]):
        lines[idx_last] = ""
        removed_line5 = True
    
    if removed_line1 and removed_line2 and removed_line5:
        lines[idx_third] = ""
        lines[idx_second] = ""
        
    text = "\n".join(line for line in lines if line.strip() != "")
    return text

for i in range(2, len(pages)):
    pages[i]['content'] = remove_footers(pages[i]['content'])
```
